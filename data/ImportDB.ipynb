```python
district = pd.read_csv("district.csv")
district.csv_path = 'district.csv'
street = pd.read_csv("street.csv")
street.csv_path = 'street.csv'
province = pd.read_csv("province.csv")
province.csv_path = 'province.csv'
domain = pd.read_csv("domain.csv")
domain.csv_path = 'domain.csv'
ranking = pd.read_csv("ranking.csv")
ranking = ranking[['hotel_id', 'rank_score']]
ranking.csv_path = 'ranking.csv'
info = pd.read_csv("info.csv")
info.csv_path = 'info.csv'
quality = pd.read_csv("quality_new.csv")
quality.csv_path = 'quality.csv'
review = pd.read_csv("review.csv")
review.csv_path = 'review.csv'
hotel = pd.read_csv("root_update_price.csv")
hotel.drop(["logo", "check_in_time", "check_out_time", "description", "min_price_domain"], axis=1, inplace=True)
hotel.csv_path = 'hotel.csv'
url = pd.read_csv("url_new.csv")
url.csv_path = 'url.csv'
```

```python
database = [district, street, province, domain, ranking, info, quality, review, hotel, url]
```

```python
for df in database:
    for column in df.columns:
        if "Unnamed: 0" in column:
            df.drop(column, axis = 1, inplace=True)
    df.to_csv(df.csv_path, sep = ',', index= False)
```
